**2.2 : Streaming en Temps Réel avec Azure Databricks et Event Hubs**

- Configurer un Event Hub pour simuler l’envoi de données en continu.
- Lire les flux de données depuis Event Hubs avec Spark Structured Streaming dans Azure Databricks.
- Sauvegarder les données transformées dans une Delta Table.
​

```python
df_kafka.printSchema()
```

```python
# Save raw Kafka data to Bronze Delta Table
df_kafka.writeStream \
    .format("delta") \
    .option("checkpointLocation", "/mnt/delta/checkpoints/bronze_kafka_2.22") \
    .option("path", "/mnt/delta/bronze_kafka_2.22") \
    .outputMode("append") \
    .start()
```

*Conclusion: Bronze Layer stores raw unprocessed Kafka messages as Delta Table.*

```python
# Define schema for JSON parsing (Silver Layer)
schema = StructType([
    StructField("vendorID", StringType(), True),
    StructField("tpepPickupDateTime", StringType(), True),
    StructField("tpepDropoffDateTime", StringType(), True),
    StructField("passengerCount", IntegerType(), True),
    StructField("tripDistance", DoubleType(), True),
    StructField("puLocationId", StringType(), True),
    StructField("doLocationId", StringType(), True),
    StructField("rateCodeId", StringType(), True),
    StructField("paymentType", StringType(), True),
    StructField("fareAmount", DoubleType(), True),
    StructField("extra", DoubleType(), True),
    StructField("mtaTax", DoubleType(), True),
    StructField("improvementSurcharge", StringType(), True),
    StructField("tipAmount", DoubleType(), True),
    StructField("tollsAmount", DoubleType(), True),
    StructField("totalAmount", DoubleType(), True)
])
```

```python
# Read and transform Bronze data into Silver Layer
df_silver = (spark.readStream
    .format("delta")
    .load("/mnt/delta/bronze_kafka_2.22")
    .selectExpr("CAST(value AS STRING) as json")
    .select(from_json(col("json"), schema).alias("data"))
    .select("data.*")
    .withColumn("tpepPickupDateTime", to_timestamp(col("tpepPickupDateTime") / 1000))
    .withColumn("tpepDropoffDateTime", to_timestamp(col("tpepDropoffDateTime") / 1000))
    .filter(col("tripDistance") > 0)  # Remove trips with zero distance
    .filter(col("totalAmount") > 0)  # Remove invalid fares
)

df_silver.display()  # Show Silver Layer data

# Save transformed data to Silver Delta Table
df_silver.writeStream \
    .format("delta") \
    .option("checkpointLocation", "/mnt/delta/checkpoints/silver_kafka_2.22") \
    .option("path", "/mnt/delta/silver_kafka_2.22") \
    .option("mergeSchema", "true") \
    .outputMode("append") \
    .start()
```

*Conclusion: Silver Layer applies schema, filters invalid records, and formats timestamps.*

```python
# Aggregated Data for Gold Layer
df_gold = df_silver.groupBy("vendorID").agg(
    expr("count(*) as total_trips"),
    expr("avg(tripDistance) as avg_distance"),
    expr("avg(totalAmount) as avg_fare"),
    expr("sum(totalAmount) as total_revenue")
)

df_gold.display()  # Show Gold Layer data

# Save analytics data to Gold Delta Table
df_gold.writeStream \
    .format("delta") \
    .option("checkpointLocation", "/mnt/delta/checkpoints/gold_kafka_2.22") \
    .option("path", "/mnt/delta/gold_kafka_2.22") \
    .outputMode("complete") \
    .start()
```

*Conclusion: Gold Layer aggregates trip data by vendor, calculating total trips, distance, and revenue.*
